```python
with open('input.txt') as f:
    contents = f.read()
    print(contents)
```

```python
newdic = dict()
string = contents
newlist = [string[start:start+6] for start in range(0, len(string), 6)]
print("layers",newlist)

for ele in newlist:
    count = ele.count('0')
    #print(ele, count)
    intcount = int(count)
    newdic[intcount] = int(ele)
    
print("counting number of 0 in each layer and selecting the one with min 0")
print(newdic)
         
newdic.items()
layer = min(newdic.items(), key=lambda x: x[0])
print("Selected Layer")
print(layer)
```
